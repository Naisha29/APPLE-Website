# UNDER **SAMPLING**

```python
build a sample dataset containing similar distribution of normal transactions and fradulent transactions
jaisa there are 492 fraud so we take 492 from legit also making it equal sample dataset
```

```python
legit_sample=legit.sample(n=492)
```

this is to select 492 out of 284315 legit items

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

Concatenating two DataFrame if axis=0 means next data added row wise if axis=1 data added is column wise

```python
new_dataset.head()
```

u see in head which gives top 5 datarows is 0 meaning legit and then all last ones are 1 meaning fraudulent

```python
new_dataset.groupby('Class').mean()
```

```python
X= new_dataset.drop(columns='Class', axis=1)
Y= new_dataset['Class']
print(X)
print(Y)
```

splitting data into features(x i.e v1,v2,v3..) and target(y i.e class 0 and 1)

Split the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

text size =0.2 meaning 20% goes to training data and 80% goes to testing data

```python
print(X.shape,X_train.shape, X_test.shape)
```

x was data set of 984 then to split in 20:80 we have x_train and x_test

model training using logisitic regression for binary classification

```python
model= LogisticRegression()
```

training the logistic model regression with training data

```python
model.fit(X_train,Y_train)
```

fit this data in logistic regression model

# **Model** **Evaluation**
accuracy score on training data


```python
X_train_prediction= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction,Y_train)
print('Accuracy on training data: ', training_data_accuracy)
```

accuracy score on test data

```python
X_test_prediction= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction,Y_test)
print('Accuracy on test data: ', test_data_accuracy)
```

accuracy of test data is more important than training data
